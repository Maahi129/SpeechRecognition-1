```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style = 'whitegrid')
plt.figure(figsize=(18,6))
f = 40
tmin = -0.3
tmax = 0.3
t = np.linspace(tmin, tmax, 400)
n = np.cos(2*np.pi*t) + np.cos(2*np.pi*f*t)
plt.plot(t, n, label = 'Frequency (40 hz)')
plt.legend()

T = 1 / 80
nmin = np.ceil(tmin/T)
nmax = np.floor(tmax/T)
n = np.arange(nmin,nmax)
x = np.cos(2*np.pi*n*T) + np.cos(2*np.pi*n*f*T)
plt.plot(n*T, x, 'ro', markersize = 7, label = 'Sampling with 80 hz')
plt.legend()
plt.show()

```
